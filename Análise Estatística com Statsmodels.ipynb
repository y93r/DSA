# Regressão Linear Simples

## Construção do Modelo OLS (Ordinary Least Squares) com Statsmodels em Python

```python
df_dsa.head()
```

```python
# Definimos a variável dependente
y = df_dsa["valor_aluguel"]
```

```python
# Definimos a variável independente
X = df_dsa["area_m2"]
```

```python
# O Statsmodels requer a adição de uma constante à variável independente
X = sm.add_constant(X)
```

```python
# Criamos o modelo
modelo = sm.OLS(y, X)
```

```python
# Treinamento do modelo
resultado = modelo.fit()
```

```python
print(resultado.summary())
```

```python
# Plot
plt.figure(figsize = (12, 8))
plt.xlabel("area_m2", size = 16)
plt.ylabel("valor_aluguel", size = 16)
plt.plot(X["area_m2"], y, "o", label = "Dados Reais")
plt.plot(X["area_m2"], resultado.fittedvalues, "r-", label = "Linha de Regressão (Previsões do Modelo)")
plt.axhspan(ymin=0, ymax=max(y), facecolor='lightblue', alpha=0.3)
plt.legend(loc = "best")
plt.show()
```

## Conclusão

Embora seja evidente que existe uma correlação entre a área dos imóveis e o valor do aluguel, é importante destacar que a área sozinha não é suficiente para explicar completamente a variação no valor do aluguel. Isso é evidenciado pelo coeficiente de determinação (R²) de apenas 0,34 obtido pelo nosso modelo de regressão linear simples.

Para obter uma compreensão mais precisa  dos fatores que influenciam o valor do aluguel, deveríamos incorporar outras variáveis explicativas ao modelo, aplicar análises causais para investigar relações de causa e efeito e validar as suposições do modelo antes de tirar conclusões definitiva.
